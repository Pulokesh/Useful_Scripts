## How to Generate SSH Keys on XShell 7, and use it to login to Niagara cluster on Compute Canada

* [1] Open XShell, and on clicking Niagara (or any other Compute Canada Clusters, such as Gaham for example here) you will get a window like this.
* [2] Click on `Public Key`, then from `Browse...` click on `User Keys...` like the one shown below.

* [3] Now you click on `Generate...` to generate new Keys. 


* [4] Next you will see `New User Key Wizard` where you select Key Type `RSA` and Key Length `4096 bits` (for example), and then you click on `Next` to create public Key.

* [5] Next, on `Generate Public Key Pair` 4096-bit key pair will be generate. Click `Next` to go to the next step.

* [6] Now, on the `User Key Information` window, enter `Key Name` and `Passphrase` and confirmation of passphrase. Note down the `Key_name` and `Passphrase`. These two will be like login information whenever you want to login to Niagara. Click `Next` to proceed to the next step.

* [7] Next, Your Register Public Key will be shown. The Key will be in `SSH2-OpenSSH` format. You can `Save as a file` and open the file and copy the content of the file which will laber be pasted on `Compute Canada` account key section

* You could directly copy the Public Key from here (which is what I actually do), to be pasted on CCDB `Add key` section.

* [8] Now, you have to login to your CCDB (Compute Canada Database) account.

* [9] After you login, go to `My Account` and `Manage SSH Keys` section.

* [10] Here, you have to `Paste Your SSH Key` in the box and then click on `Add New`. Finally, you are done with adding an SSH key.

## Login with an SSH Key from XShell 7
* Everytime you want to login to Niagara you have to enter `User Key` and `Passphrase`. The User Key is stored on XShell, but you have to remember the `Passphrase`. From now on, you dont need password to login to Niagara.
